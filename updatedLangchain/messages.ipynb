```python
import os
from langchain.chat_models import init_chat_model

groq_api_key = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:llama-3.3-70b-versatile")
```

```python
model.invoke("hello how are you ..")
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage  

messages=[
    SystemMessage("You are a poetry expert"),
    HumanMessage("write a poem on artificial intelligence ")
]
response = model.invoke(messages)
response.content
```

```python
sys_msg = SystemMessage("""
    you are a coding helpful assisstant, specialization with senior python develpoer and lots of knowldege in coding field  
""")

messages=[
    sys_msg,
    HumanMessage("how do i create api ")
]

data = model.invoke(messages)
print(data.content)
```

```python
from langchain_core.messages import (
    HumanMessage,
    AIMessage,
    ToolMessage
)

messages = [

    HumanMessage(
        content="What's weather in Delhi?"
    ),

    AIMessage(
        content="",
        tool_calls=[
            {
                "name": "get_weather",
                "args": {
                    "location": "Delhi"
                },
                "id": "tool_1"
            }
        ]
    ),

    ToolMessage(
        content="It's sunny in Delhi",
        tool_call_id="tool_1"
    )
]

print(messages)
```
