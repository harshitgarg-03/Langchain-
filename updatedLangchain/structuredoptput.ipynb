```python
import os
from langchain.chat_models import init_chat_model

groq_api_key = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:llama-3.3-70b-versatile")
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title:str=Field(description="The title of the movie ")
    year:int=Field(description="THis year the movie was released")
    director:str=Field(description="The director of the movie ")
    rating:float=Field(description="the rating of this movie out of 10")
```

```python
model_with_structure = model.with_structured_output(Movie)

print(model_with_structure)
```

```python
# THis is without structured data 
model.invoke("Provide the deatils of movie speacial26")
```

```python
# THis is structured data 
response = model_with_structure.invoke("Provide the deatils of movie speacial26")
print(response)
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    """A movie with details """
    title:str=Field(description="The title of the movie ")
    year:int=Field(description="THis year the movie was released")
    director:str=Field(description="The director of the movie ")
    rating:float=Field(description="the rating of this movie out of 10")

model_with_structure = model.with_structured_output(Movie, include_raw=True)
response = model_with_structure.invoke("Provide the deatils of movie speacial26")
print(response)
# in thsi ai message also include with structured output because of inclueraw
```

Nested Structure 

```python
from pydantic import BaseModel, Field

class Actor(BaseModel):
    name:str
    role:str

class MovieDetails(BaseModel):
    """A movie with details """
    title:str=Field(description="The title of the movie ")
    year:int=Field(description="THis year the movie was released")
    director:str=Field(description="The director of the movie ")
    rating:float=Field(description="the rating of this movie out of 10")
    cast:list[Actor]
    genres:list[str]
    budget:float | None = Field(None, description="THe budget in millions with usd ")

model_with_structure = model.with_structured_output(MovieDetails)
response = model_with_structure.invoke("Provide the deatils of movie speacial26")
print(response)
```

TYpeDict

```python
# Basically typedict not validations of field while Pydantic validations of field 

from typing_extensions import Annotated, TypedDict

class MovieDetailsWithTypeDict(TypedDict):
    """A movie with details """
    title:Annotated[str, ..., "The title of the movie "]
    year:Annotated[int, ..., "The year of the movie "]
    director:Annotated[str, ..., "The director of the movie "]
    rating:Annotated[float, ..., "The rating of the movie "]
    budget:Annotated[str, ..., "The budget of the movie in millions "]

model_with_structure = model.with_structured_output(MovieDetailsWithTypeDict)
response = model_with_structure.invoke("Provide the deatils of movie speacial26")
print(response)
```

```python
# pydantic based 

from langchain.agents import create_agent
from pydantic import BaseModel, Field


class Contact(BaseModel):
    """A contact with details"""

    PersonName: str = Field(
        description="The name of the person"
    )

    PhoneNum: str = Field(
        description="The phone number of the person"
    )

    CountryCode: str = Field(
        description="The country code of phone number"
    )

    Email: str = Field(
        description="The email of the person"
    )


agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    response_format=Contact
)


result = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": 
            "Extract contact info:Harshit Garg,+91 7819007832,garg123@gmail.com"
        }
    ]
})

# print(result)

result["structured_response"]
```

```python
# dataclasses based

from dataclasses import dataclass
from langchain.agents import create_agent

@dataclass
class ContactInfo(BaseModel):
    """A contact with details"""

    PersonName: str = Field(
        description="The name of the person"
    )

    PhoneNum: str = Field(
        description="The phone number of the person"
    )

    CountryCode: str = Field(
        description="The country code of phone number"
    )

    Email: str = Field(
        description="The email of the person"
    )

agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    response_format=ContactInfo
)

result = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": 
            "Extract contact info:Harshit Garg,+91 7819007832,garg123@gmail.com"
        }
    ]
})

result["structured_response"]
```
